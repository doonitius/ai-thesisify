```python
import nltk
import string
import pandas as pd
from collections import defaultdict
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import snowball
from pprint import pprint
from glob import glob, glob1 #FROM EXTRACT PDF
nltk.download('punkt')
nltk.download('stopwords')
stemmer = snowball.SnowballStemmer('english')
stwords = set(stopwords.words('english'))
```

```python
path = './txtFile/'
inverted_index = defaultdict(int)
word_array = []
words_count = defaultdict(int)
word_doc_count = defaultdict(int)
df_list = []
tf_list = []
idf_list = []
tf_idf_list = []
id = 1

file_count = 0
check_file_count = 0

data = pd.DataFrame(columns=['id', 'keywords', 'count'])
tf_words = pd.DataFrame(columns=['id', 'keywords', 'tf'])
idf_words = pd.DataFrame(columns=['keywords', 'idf'])
tf_idf_words = pd.DataFrame(columns=['id', 'keywords', 'tf_idf'])
```

```python
def add_data_into_df(id, data):
    for count, val in enumerate(inverted_index):
        # print(id, val)
        data = data.append({"id": id, "keywords": list(inverted_index)[count], "count": inverted_index.get(val)}, ignore_index=True)
    return data
```

```python
def read_all_words (f, doc_id):
    raw_text = f.read()
    tokens_count = nltk.word_tokenize(raw_text)
    words_count[doc_id] = len(tokens_count)
    return
```

```python
for i in glob1('./txtFile/', '*'):
    with open(path + i, 'r', encoding = 'utf8') as f:
        doc_id = i.split('.')[-2]
        doc_id_int = int(doc_id)
        file_count += 1
        read_all_words(f, doc_id)
    f.close()
```

```python
for i in glob1('./txtFile/', '*'):
    with open(path + i, 'r', encoding = 'utf8') as f:
        doc_id = i.split('.')[-2]
        doc_id_int = int(doc_id)
        check_file_count += 1
        if id != doc_id_int:
            data_new = add_data_into_df(id, data)
            df_list.append(data_new)
            id = doc_id_int
            inverted_index.clear()
        for j in f:
            line = j.strip()
            for sent in sent_tokenize(line):
                for word in word_tokenize(sent):
                    word_lower = word.lower()
                    if word_lower not in stwords:
                        word_stem = stemmer.stem(word_lower)
                        if not word_stem.isdigit():
                            punct_detected = False  
                            for x in word_stem:
                                if x in string.punctuation:
                                    punct_detected = True
                            if punct_detected is False:
                                if word_stem in inverted_index.keys():
                                    inverted_index[word_stem] = inverted_index.get(word_stem) + 1
                                else:
                                    inverted_index[word_stem] = 1
                                if len(word_array) == 0:
                                    word_array.append(word_stem)
                                else:
                                    if word_stem in word_array:
                                        break
                                    else: word_array.append(word_stem)
        if check_file_count == file_count:
            data_new = add_data_into_df(id, data)
            df_list.append(data_new)
            id = doc_id_int
            inverted_index.clear()
    f.close()
```

```python
hold_i = 0
check_in_loop = 0
```

```python
#TF
for i in range(0, file_count):
    # print(file_count)
    # print(i)
    df = df_list[i]
    id = df['id'][0]
    # print(id)
    count = words_count.get(str(id))
    # print(id, hold_i)
    if id != hold_i:
        tf_list.append(tf_words)
        hold_i = id
        tf_words = pd.DataFrame(columns=tf_words.columns)
    for j in df.index:
        tf = df['count'][j] / count
        # print(id,df['keywords'][j],tf)
        tf_words = tf_words.append({"id": id, "keywords": df['keywords'][j], "tf": tf}, ignore_index=True)
    check_in_loop += 1
    if check_in_loop == file_count:
        tf_list.append(tf_words)
        hold_i = id
        tf_words = pd.DataFrame(columns=tf_words.columns)
```

```python
import math

for i in range(0, file_count):
    df = df_list[i]
    id = df['id'][0]
    for j in df.index:
        if df['keywords'][j] in word_doc_count.keys():
            word_doc_count[df['keywords'][j]] = word_doc_count.get(df['keywords'][j]) + 1
        else:
            word_doc_count[df['keywords'][j]] = 1
    
```

```python
keys_list = list(word_doc_count)
```

```python
# IDF
# IDF(t)  = log [ (1 + n) / (1 + DF(t)) ] + 1
for i in range(len(word_doc_count)):
    idf = math.log((1+file_count) / (1+word_doc_count[keys_list[i]])) + 1
    idf_words = idf_words.append({"keywords": keys_list[i], "idf": idf}, ignore_index=True)
idf_list.append(idf_words)
```

```python
# TF-IDF
# tf_list.pop(0) # how to do it very time we run
# tf_list[0]

for i in range(len(idf_list[0])):
    for j in range(len(tf_list)):
        for k in tf_list[j].index: 
            keyword = tf_list[j]['keywords'][k]
            id = tf_list[j]['id'][k]
            if keyword == idf_list[0]['keywords'][i]:
                tf_idf = tf_list[j]['tf'][k] / idf_list[0]['idf'][i]
                tf_idf_words = tf_idf_words.append({"id": id, "keywords": keyword, "tf_idf": tf_idf}, ignore_index=True)
                # print(tf_idf_words)
```
